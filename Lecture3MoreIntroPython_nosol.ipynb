# Variables and assignment

Programs would be unreadable if they all consisted of gigantic expressions.  A common way to logically break things up is to assign an expression's value to a variable.


````python
a = 4    # Assign the value 4 to the variable a
print(a + 2)
````

The righthand side of = can also be an expression.

````python
a = 2 + 2
b = 7 * 2
print(a+b)
````

Variable names ("identifiers") are best when they're both short and descriptive.  If you use multiple words for a variable name, separate them with an underscore _ as a matter of convention.  They can't start with digits.  Don't use a name already in use, like a function name or a Python keyword.

````python
dollars_per_hour = 50
hours_worked = 40
total_dollars = dollars_per_hour * hours_worked
total_dollars
````

Any variable can get any data type assigned to it.  The four "primitive" data types in Python that don't combine other data types are:

* int - integer
* float - number with a decimal point
* str - string
* bool - boolean (True or False)

It's good to pick a variable name that will also remind you of the type you stored in it.  What do you think would be stored in days_until_vacation?  currently_on_vacation?  hotel_name?  cost_per_night?

# Branching execution

Branching refers to a situtation where if a condition holds, one thing will happen, and if not, a different thing will happen.  Branching is mostly handled by "if" statements.  Their basic structure is

```
if condition:
  code line 1
  code line 2
  ...
  code line n
else:
  code line 1'
  code line 2'
  ...
  code line n'

```


The condition is an expression that evaluates to a Boolean, like "a < 5" to check whether variable *a* has a value less than 5.  The colon at the end of the if is mandatory to signal that what follows is the stuff to execute if the condition is true.  Then, the optional else with its colon signals that the block to follow is what to do if the condition evaluated to false.

````python
name = input('Enter name:'); # input() gets user input; argument is prompt
if len(name) > 0:
    print("Name: " + name)
    print("Welcome!")  # Also indented, so only executes in len > 0 branch
else:
    print("No name")
print("Done!")  # Not indented, so always happens after if/else
````

One variation on the if/else branching is that the "else" isn't strictly necessary.  You could just have some code that only happens if the condition is true, and if the condition isn't true, execution just keeps going.

````python
a = int(input('Enter a number:'))     # Let's assume the input is an integer
if a == 789:
    print("Why is 6 afraid of 7?")
    print("... Because 7 regularly puts on scary clown makeup ")
    print("and listens to music backwards at midnight?")
    print("I was not aware of that.  I guess that's a good answer.")
    print("But also ...")
print(a)
````

It's possible you need a branch to go multiple ways, not just a two-option this or that.  In that case, you can handle multiple branches with "if" for the first option, "elif" for the second through second-to-last options, and "else" to handle everything else.

````python
state = input('Enter state: ')
if state == "California":
    print('It is illegal to be uptight in the current state.')
elif state == "Massachusetts":
    print('It is illegal to act like a Californian after 10pm in the current state.')
elif state == "New Jersey":
    print('Nothing is illegal in the current state.')
else:
    print('It is illegal to break the law in the current state.')
````

# Boolean Operators

Sometimes the condition for some code executing is more complex than just one condition.

* a **and** b returns True only if a and b are *both* True.  (Use if there are *multiple requirements* to execute.)
* a **or** b returns True only if *at least one* of a and b is True.  (Use if *any of a few conditions* are sufficient to execute.)
* **not** a returns True only if a is False.  (Use if it's easier to say when *not* to execute.)

And happens before or

````python
my_money = 10
debts = -10
vip = False
if my_money >= 0 and my_money <= 1000000:
    print('my_money is in a valid range')
if my_money < 0 or debts < 0:
    print('Error - was expecting all values to be non-negative')
if not vip:
    print('Consider upgrading to our premium version!')
````

We can combine boolean operators in the same expression.  Suppose I want to set a variable "important_client" that is always true if the "vip" variable is true, but can still be true if my_money is at least a million and debts is 0.  (It's False if vip is False and one of those other conditions isn't met.)  How could we write a single line that does this?

````python
imortant_client = vip or (my_money >= 1000000 and debts == 0)
````

# Lists

A list is a *data structure*, which is to say, it's a data type that holds data of other types.  A list can be a list of strings, or a list of numbers, or even a list of lists.  It can even hold multiple data types at once, like strings and numbers, although typically every element is of the same type.

You can create a list by putting comma-separated values between square brackets, like so:

````python
client_names = ["Alice", "Beatrice", "Connie"]
contract_values = [100000, 200000, 300000]
````

To retrieve an individual item from the list, give the list name, followed by the index of the item in square brackets.  For historical reasons, the counting starts from 0.

````python
client_names = ["Alice", "Beatrice", "Connie"]
print(client_names[0])
print(client_names[1])
print(client_names[2])
````

You can also assign values to a list using this "array notation."  When it is used on the left hand side of assignment, the value at that place is replaced with the right-hand side value.

````python
client_names[1] = "Bob"
print(client_names)
````

If you need to get to an item inside a list that itself is a list - a list-of-lists - you can index and then index again, like so:

````python
client_value_histories = [
     [30000,20000,20000],
     [10000,20000,30000],
     [50000,50000,60000]]
print(client_value_histories[0]) # The first client's value history
print(client_value_histories[0][1]) # 2nd value of 1st list
````

A simple way to concatenate (stick together) two lists is to use the + operator, similar to how strings treat the operator.

````python
clients = ['Alice', 'Bob', 'Cordelia']
more_clients = ['Daphne', 'Eliot']
print(clients + more_clients)
````

There are other data structures that we'll cover besides the list, but Python's list implementation makes it very flexible as an all-purpose container of data.  You can look at the list documentation at https://docs.python.org/3/tutorial/datastructures.html to see the many built-in functions ("methods") that lists have.

.append is a method


````python
clients = ['Alice', 'Bob', 'Cordelia']
clients.append('Daphne') # Now doing business with Daphne
print(clients)
# Add Berenice, but insert at position 1 to maintain
# alphabetical order
clients.insert(1, 'Berenice')
print(clients)
print('Berenice now at pos ' + str(clients.index('Berenice')))
# Here's another way to keep things sorted:  sort()
clients.append('Aaron')
clients.sort()
print(clients)
````

# Median

The **median** is an important metric in data science.  It is the middle value if all the values were lined up from smallest to biggest.  (Or if the number of items is even, the average of the two middle values.)  Unlike the mean (average), the median is less influenced by extreme values (outliers) and gives a sense of what a "typical" value is.

How could we use what we've covered so far about lists to easily find the *median* value in, say, a list of 9 planet diameters (including Pluto to make the number odd)?

````python
# Diameters in order from Sun:  Mercury, Venus, Earth, Mars...
planet_diameter_km = [4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528, 2377]
````

````python
planet_diameter_km.sort()
print(planet_diameter_km[4]) # "Hard-coding" the 5th place of 9
# More general solution follows for odd-sized lists.
# Note that we can't index lists with floating point numbers,
# which division gives, so we must cast to an integer
median_index = int((len(planet_diameter_km)-1)/2)
print(planet_diameter_km[median_index])
````
