# Strings

Expressions can also include text.  The technical programming term for a snippet of text is a "string."  In Python, strings are surrounded by either single quotes or double quotes, thus 'this is a string' or "this is a string".  Python considers strings to be very different from numbers, so '2' + '2' may not do what you expect.


```python
'2' + '2'
```

The + operator happens to work with strings, but it no longer means addition.  It means *concatenation*, or joining two strings together to create one string.

```python
'hello,' + ' world!'
```

A number can be converted to a string in the following way.  This is calling a function, which we'll say more about later.


```python
str(3) + " little pigs"
```

It's quite common to want to report some numerical value computed by your program within some text that makes it clear what the number means.  But to use concatenation, first the number must be converted to a string.  Converting between one type of data and another is called *casting*.  We are casting a number to a string.  And if we don't do that ...

```python
"the " + 3 + " amigos"
```

Python error messages are sometimes hard to understand, but here the error message is relatively clear:  Python does not want to concatenate data of different types, and we tried to concatenate an integer (int) and a string (str).  Also notice that an arrow points to where the error happened - useful when debugging big programs.

# Calling functions

We've so far seen two functions in this document:  print(), which prints a value to the screen, and str(), which converts a number to a string so that it can be concatenated with other strings.

A function can have zero or more arguments, or inputs -- usually it takes at least one -- and then evaluates to something - it *returns* a value.  Let's take a look at *len()* as an example, which takes a string and returns the length in characters.

```python
len("Go BU")
```

```python
len("Go") + len("Terriers")
```

Functions can be a part of expressions, and we can also use expressions as function arguments.

```python
print('"Go BU" is ' + str(len('Go BU')) + ' characters long')
```

Not all functions evaluate to what you'd expect.  print() just has the "side effect" of printing to the screen, but it returns the special value None, meaning "this function doesn't return anything."

```python
print("Crash") + " into me"
```

Functions are organized into modules that need to be imported before their functions can be used.  Just for this example, we'll import the math module.  This is necessary to use any functions in that module.

```python
import math
```

Now we can call any function in the math module, as long as we preface it with "math."  We can take square roots, for example.

```python
math.sqrt(4)
```

The documentation at https://docs.python.org/3/library/math.html tells us that math contains pow(), which raises its first argument to the power of its second argument.  There happens to also be a Python operator for this, **.

```python
math.pow(9,2)
```

```python
9 ** 2
```

# Code comments

A comment is some text written after a # sign, which Python ignores as it executes.  Good programmers leave comments in their code to help their human readers understand it.

```python
my_money = 10  # Contains total assets - total debts of user
```

You don't need to comment anything obvious, but it's common to comment anything unusual, and also provide the reader with a high-level idea of what a block of code does.  It may help you make sense of your own code later, too.

```python
# Print the prime numbers between 2 and 31.
for i in range(2,32):
    prime = True
    for j in range(2,i):  # Could make this more efficient
        if i % j == 0:
            prime = False
            break
    if prime:
        print(i)
```

It's also common to put long comments in multiline strings, as shown below.  These start and end with triple double-quotes: """.

```python
"""Sorry for the complicated code that follows.
It's because I'm a very complicated person.
How complicated?  Well, ever seen the movie Primer?
Yeah.  WAY more complicated than that."""
```

*Cynthia contemplates moving to New York to pursue a job doing data science in finance.  But how far away is that from Boston, exactly?  She decides to flex her new Python skills and figure it out.*

New York City is at 40.7128° N, 74.0060° W.  Boston is at 42.3601° N, 71.0589° W.  But we can't just use our formula for Euclidean distance on these coordinates, because latitude and longitude convert to different numbers of miles.  One degree of latitude is 69 miles.  One degree of longitude's conversion to miles depends on where you are on the globe, but at 40 degrees North it's 53 miles.

```python
(42.3601-40.7128) * 69  # Latitude difference, converted to miles
```

```python
(74.0060-71.0589) * 53 # Longitude difference, converted to miles
```

```python
import math
# Euclidean distance formula; approximation on a sphere!
math.sqrt(113.6637 ** 2 + 156.1963 ** 2)
```
